# Capitec Data Science Course — Load Dataset

This notebook loads all CSV files from the `data/` folder of this repository into pandas DataFrames.

It works two ways:
- **If you cloned the repo** (GitHub Desktop or `git clone`), it will read the files locally from `../data/`.
- **If you only downloaded this notebook**, it falls back to reading the files directly from GitHub (raw URLs), since the repository is public.

```python
import os
import pandas as pd

# Local path (when repo is cloned) vs. GitHub raw URL (fallback)
LOCAL_DATA_DIR = "../data/"
RAW_BASE_URL = "https://raw.githubusercontent.com/Geeks4LearningJHB/capitec-data-science-test-repo/main/data/"

FILES = [
    "dim_campaign.csv",
    "dim_channel.csv",
    "dim_client.csv",
    "dim_date.csv",
    "dim_geography.csv",
    "dim_industry.csv",
    "dim_month.csv",
    "dim_product.csv",
    "fact_campaign_touch.csv",
    "fact_client_monthly.csv",
    "fact_digital_events.csv",
    "fact_product_adoption.csv",
]
```

```python
def load_dataset(filename: str) -> pd.DataFrame:
    """Load a CSV from the local data/ folder if present, otherwise from GitHub."""
    local_path = os.path.join(LOCAL_DATA_DIR, filename)
    if os.path.exists(local_path):
        return pd.read_csv(local_path)
    return pd.read_csv(RAW_BASE_URL + filename)


data = {f.replace(".csv", ""): load_dataset(f) for f in FILES}

print(f"Loaded {len(data)} tables:")
for name, df in data.items():
    print(f" - {name}: {df.shape[0]} rows x {df.shape[1]} columns")
```
